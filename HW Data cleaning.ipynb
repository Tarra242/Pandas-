```python
#Question 1 How many flights were there from JFK to SLC? Int
# q_1

num_flights = len(df[(df["origin"] == "JFK") & (df["dest"] == "SLC")])
num_flights
```

```python
# Question 2 How many airlines fly to SLC? Should be int
# q_2

q_2 = df[df["dest"] == "SLC"]["carrier"].nunique()  # dest filters to SLC and carrier picks airline column and unique is how many airlines
q_2
```

```python
# Question 3 What is the average arrival delay for flights to RDU? float
# q_3

q_3 = df[df["dest"] == "RDU"]["arr_delay"].mean() # This filters flights heading to RDU and calculates the average arrival delay
q_3
```

```python
# Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)? float
# q_4
total_sea = len(df[df["dest"] == "SEA"])
nyc_sea = len(df[(df["dest"] == "SEA") & (df["origin"].isin(["LGA", "JFK"]))]) # flights from LGA/JFK
q_4 = nyc_sea / total_sea
q_4 * 100   # this creates an output as a % 
```

```python
# Question 5 Which date has the largest average depature delay? Pd slice with date and float please make date a column. 
# Preferred format is 2013/1/1 (y/m/d)
# q_5

df["date"] = df["year"].astype(str) + "/" + df["month"].astype(str) + "/" + df["day"].astype(str) # correct formatting 

avg_delay_by_date = df.groupby('date')['dep_delay'].mean().reset_index()  # grouped by date and mean departure delay

q_5 = avg_delay_by_date.loc[avg_delay_by_date["dep_delay"].idxmax()] # largest mean delay 
q_5
```

```python
# Question 6 Which date has the largest average arrival delay? pd slice with date and float
# q_6

avg_arr_delay = df.groupby('date')['arr_delay'].mean().reset_index() # filtered by date and average delay 
q_6 = avg_arr_delay.loc[avg_arr_delay["arr_delay"].idxmax()]
q_6
```

```python
# Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime
# q_7

df_2013 = df[(df['year'] == 2013) & (df['origin'].isin(['LGA', 'JFK']))] # correct subsets? 

df_2013['speed'] = df_2013['distance'] / (df_2013['air_time'] / 60)  # calculated in mph 

q_7 = df_2013.loc[df_2013['speed'].idxmax(), ['tailnum', 'speed']] #flight with highest speed
q_7
```
